```python
import csv, json

rows = []
titles = set()
with open('/Users/shubhamarundekatey/Downloads/OFLC_Wages_2025-26_Updated/merged.csv') as f:
    reader = csv.DictReader(f)
    for row in reader:
        titles.add(row['OES_SOCTITLE'])
        rows.append([
            row['CountyTownName'],
            row['StateAb'],
            row['OES_SOCTITLE'],
            row['Level1'],
            row['Level2'],
            row['Level3'],
            row['Level4'],
            row['Average']
        ])

out = {'titles': sorted(titles), 'rows': rows}
print(json.dumps(out, separators=(',', ':')))
```
