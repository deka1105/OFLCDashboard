```python
ALC_df = pd.read_csv("/Users/shubhamarundekatey/OFLCDashboard/OFLC_Wages_2025-26_Updated/ALC_Export_26.csv")
Geo_df = pd.read_csv("/Users/shubhamarundekatey/OFLCDashboard/OFLC_Wages_2025-26_Updated/Geography.csv")
Pro_df = pd.read_csv("/Users/shubhamarundekatey/OFLCDashboard/OFLC_Wages_2025-26_Updated/xwalk_plus.csv")
ALC_df['ALC_combined'] = ALC_df.apply(lambda x: '2026*' + '*'.join(x.astype(str)), axis=1)
Geo_df['Geo_combined'] = Geo_df.apply(lambda x: '2026*' + '*'.join(x.astype(str)), axis=1)
Pro_df['Pro_combined'] = Pro_df.apply(lambda x: '2026*' + '*'.join(x.astype(str)), axis=1)
```

```python
result = ALC_df.merge(Geo_df, on='Area', how='inner')
result2 = result.merge(Pro_df, left_on='SocCode', right_on='OES_SOCCODE', how='inner')
```

```python
dfx = dfx.drop(columns=['ALC_combined', 'Pro_combined', 'Geo_combined', 'Label'])
dfx.head()
```

```python
import pandas as pd
import os

df = dfx

output_dir = "/Users/shubhamarundekatey/OFLCDashboard/DividedCSV"
os.makedirs(output_dir, exist_ok=True)

for category, group in df.groupby('OES_SOCCODE'):
    filepath = os.path.join(output_dir, f'{category}.csv')
    
    if os.path.exists(filepath):
        # Update: append new rows, drop duplicates if needed
        existing = pd.read_csv(filepath)
        updated = pd.concat([existing, group], ignore_index=True)
        updated.to_csv(filepath, index=False)
    else:
        # Create: write fresh
        group.to_csv(filepath, index=False)

# Output: output/A.csv, output/B.csv, output/C.csv
```
